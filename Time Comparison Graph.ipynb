```python
import matplotlib.pyplot as plt

# Data for the bar graph
models = ['CPU', 'GPU']
run_times = [5.43, 95]  # Replace with the actual run times (e.g., in seconds)
colors = ['skyblue', 'salmon']  # Colors for the bars

# Create the bar graph
plt.figure(figsize=(8, 6))
plt.bar(models, run_times, width=0.5, color=colors)

# Adding titles and labels
plt.title('Comparison of Model Run Times', fontsize=16)
plt.xlabel('Machine Learning Models', fontsize=12)
plt.ylabel('Run Time (Hours)', fontsize=12)

# Adding values on top of the bars
for i, v in enumerate(run_times):
    plt.text(i, v + 2, str(v), ha='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
